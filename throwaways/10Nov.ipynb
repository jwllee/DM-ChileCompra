```python
dup_list = ['Nombre', 'CompradorCargoUsuario', 'CompradorCodigoOrganismo', 'CompradorCodigoUnidad',
           'CompradorCodigoUsuario', 'CompradorComunaUnidad', 'CompradorDireccionUnidad',
           'CompradorNombreOrganismo', 'CompradorNombreUnidad', 'CompradorNombreUsuario',
           'CompradorRegionUnidad', 'CompradorRutUnidad', 'CompradorRutUsuario', 'MontoEstimado', 'Descripcion']
tdup_df = tender_df.loc[tender_df.duplicated(dup_list)]
for tname in tdup_df['Nombre'].values:
    # get the tenders with this name
    tenders = tender_df[(tender_df['Nombre']==tname)]
    # check number of unique state to see if there's changes
    if tenders['Estado'].unique().shape[0] > 1:
        print('{} has states: {}'.format(tenders['CodigoExterno'].values, tenders['Estado'].unique()))
```
